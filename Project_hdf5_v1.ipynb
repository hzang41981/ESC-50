```python
import os
import csv
import shutil

#os.chdir("<abs path to desktop>")

def remake_folder(csv_filepath, old_wav_folder, new_wav_folder):
    with open(csv_filepath) as f:
        folder_name_set = set()
        f_csv = csv.reader(f)
        next(f_csv)
        for row in f_csv:
            folder_name = row[3]
            file_name = row[0]
            if folder_name not in folder_name_set:
                os.mkdir(os.path.join(new_wav_folder, folder_name))
                shutil.copy(os.path.join(old_wav_folder, file_name), os.path.join(new_wav_folder, folder_name))
                folder_name_set.add(folder_name)
            else:
                shutil.copy(os.path.join(old_wav_folder, file_name), os.path.join(new_wav_folder, folder_name))

remake_folder('/content/ESC-50-master/meta/esc50.csv', '/content/ESC-50-master/audio', '/content/ESC-50-Reorg')
```
